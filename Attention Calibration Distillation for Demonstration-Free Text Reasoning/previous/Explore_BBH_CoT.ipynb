# Explore Raw BBH Dataset
Run this cell to dynamically pull 100 random examples from Hugging Face and print the raw, unformatted dictionaries directly to the output.

```python
import random
import pprint
from datasets import load_dataset

# We will sample from 5 reasoning-heavy BBH tasks (20 from each to make 100)
tasks = [
    "boolean_expressions", 
    "navigate", 
    "snarks", 
    "logical_deduction_five_objects",
    "date_understanding"
]

dataset_name = "lukaemon/bbh"
print(f"Connecting to Hugging Face to load {dataset_name}...\n")

all_examples = []

for task in tasks:
    try:
        # Load the test split for the specific task
        dataset = load_dataset(dataset_name, task, split="test")
        
        # Randomly sample 20 examples from each task
        population_size = len(dataset)
        sample_size = min(20, population_size)
        sampled_indices = random.sample(range(population_size), sample_size)
        sampled_data = dataset.select(sampled_indices)
        
        for item in sampled_data:
            # Keep the EXACT raw dictionary from the dataset
            all_examples.append(item)
    except Exception as e:
        print(f"Error loading {task}: {e}")

# Randomize the final list of 100 examples
random.shuffle(all_examples)

print(f"\nTotal raw examples pulled: {len(all_examples)}\n")
print("=" * 80)

# Print them completely raw
for i, raw_example in enumerate(all_examples):
    print(f"[Raw Example {i+1}]")
    pprint.pprint(raw_example)
    print("-" * 80)
```
